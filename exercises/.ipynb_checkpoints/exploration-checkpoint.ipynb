```r
library(readr)
suppressMessages(library(dplyr))


game_reviews <- read_csv("../../data/ign.csv", col_types = cols())
game_reviews %>% head
```

```r
game_reviews %>% dim
```

```r
game_reviews %>% 
    group_by(title) %>% 
    summarise(meanscore = mean(score)) %>% 
    arrange(desc(meanscore)) %>% 
    head(3)
    
```

```r

game_reviews %>% 
    filter(platform == 'PC') %>% 
    arrange(desc(score),desc(release_year)) %>% 
    filter(score<9.4,score>9) 
```
